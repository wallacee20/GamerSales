```python
# Contagem total dos valores nulos
df.isnull().sum().sum()
```

```python
# Laço FOR para calcular a porcentagem de valores nulos/vazios por coluna
total_linhas = len(df)

print("Porcentagem de valores nulos/vazios por coluna:\n")

for coluna in df.columns:
    nulos = df[coluna].isnull().sum()
    vazios = (df[coluna] == '').sum()
    total_faltantes = nulos + vazios
    percentual = (total_faltantes / total_linhas) * 100
    print(f"{coluna}: {percentual:.2f}%")
```
